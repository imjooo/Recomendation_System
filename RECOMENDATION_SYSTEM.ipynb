```python
books=pd.read_csv('bookss.csv',encoding='Latin-1',error_bad_lines=False)
```

```python
books
```

```python
book=books.iloc[:,1:]
book
```

```python
book.sort_values(['User.ID'])
```

```python
len(book['Book.Title'].unique())
```

```python
books_pivot=book.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books_pivot
```

```python
# Replacing the index values by unique user Ids
books_pivot.index=book['User.ID'].unique()
books_pivot
```

```python
books_pivot.fillna(0,inplace=True)
books_pivot
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(books_pivot.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
similar_user=pd.DataFrame(user_sim)
similar_user
```

```python
# Set the index and column names to user ids
similar_user.index=book['User.ID'].unique()
similar_user.columns=book['User.ID'].unique()
similar_user
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
similar_user
```

```python
# Most Similar Users
similar_user.idxmax(axis=1)
```
